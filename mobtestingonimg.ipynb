```python
import cv2
image_path ="/content/drive/MyDrive/PlantVillage_split/test_predict/Corn___Common_rust__image (1174).JPG"
# reading img
img = cv2.imread(image_path)# when we read we get bgr format we have to convert it
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
 # Displaying img
plt.imshow(img)
plt.title("Original Image")
plt.show()
```

```python
# image in to array because neral network takes on array because model is traned on 128 128
```

```python
image= tf.keras.preprocessing.image.load_img(image_path, target_size=(128,128))
input_arr= tf.keras.preprocessing.image.img_to_array(image)
input_arr= np.array([input_arr])# covert the img into batch
input_arr.shape
```

```python
prediction= model.predict(input_arr)
prediction,prediction.shape
```

```python
result_index=np.argmax(prediction)# it gives the max index class no
result_index
```
